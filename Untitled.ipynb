```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv("dm3.kc167.tads.train.csv", header=None)
test_df = pd.read_csv("dm3.kc167.tads.test.csv", header=None)

print("Training shape:", train_df.shape)
print("Testing shape:", test_df.shape)
```

```python
train_df.head()
```

```python
X_train_seq = train_df[0]
y_train = train_df[1]

X_test_seq = test_df[0]
y_test = test_df[1]
```

```python
X_train_seq = X_train_seq.str.upper()
X_test_seq = X_test_seq.str.upper()
```

```python
print(len(X_train_seq.iloc[0]))
```

```python
mapping = {
    'A': [1,0,0,0],
    'T': [0,1,0,0],
    'G': [0,0,1,0],
    'C': [0,0,0,1],
    'N': [0,0,0,0]
}
```

```python
def encode_sequence(seq):
    return np.array([mapping.get(base, [0,0,0,0]) for base in seq])
```

```python
X_train = np.array([encode_sequence(seq) for seq in X_train_seq])
X_test = np.array([encode_sequence(seq) for seq in X_test_seq])

y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train_main, X_val, y_train_main, y_val = train_test_split(
    X_train,
    y_train,
    test_size=0.2,
    random_state=42,
    stratify=y_train
)

print("X_train_main shape:", X_train_main.shape)
print("X_val shape:", X_val.shape)
print("y_train_main shape:", y_train_main.shape)
print("y_val shape:", y_val.shape)
```
