```python
import pandas as pd
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# -----------------------------
# File paths
# -----------------------------
TRAIN_FILE = "dm3.kc167.tads.train.csv"
TEST_FILE = "dm3.kc167.tads.test.csv"

# -----------------------------
# DNA encoding
# -----------------------------
BASE_MAPPING = {
    'A': [1, 0, 0, 0],
    'T': [0, 1, 0, 0],
    'G': [0, 0, 1, 0],
    'C': [0, 0, 0, 1],
    'N': [0, 0, 0, 0]
}

def encode_sequence(seq, mapping=BASE_MAPPING):
    return np.array([mapping.get(base, [0, 0, 0, 0]) for base in seq], dtype=np.float32)

def load_dataset(csv_file):
    df = pd.read_csv(csv_file, header=None)
    sequences = df[0].astype(str).str.upper().values
    labels = df[1].values
    X = np.array([encode_sequence(seq) for seq in sequences], dtype=np.float32)
    y = np.array(labels)
    return X, y

# -----------------------------
# Model builder
# -----------------------------
def build_cnn(
    input_shape,
    conv_filters=(32, 64, 64),
    kernel_size=15,
    dense_units=64,
    dropout_rate=0.5,
    learning_rate=0.001
):
    model = Sequential()
    model.add(Input(shape=input_shape))

    for filters in conv_filters:
        model.add(Conv1D(filters, kernel_size, activation='relu', padding='same'))
        model.add(MaxPooling1D(2))

    model.add(Flatten())
    model.add(Dense(dense_units, activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

# -----------------------------
# Training / evaluation helper
# -----------------------------
def run_experiment(
    name,
    X_train,
    y_train,
    X_test,
    y_test,
    conv_filters=(32, 64, 64),
    kernel_size=15,
    dense_units=64,
    dropout_rate=0.5,
    learning_rate=0.001,
    patience=5,
    epochs=30,
    batch_size=32
):
    model = build_cnn(
        input_shape=(X_train.shape[1], X_train.shape[2]),
        conv_filters=conv_filters,
        kernel_size=kernel_size,
        dense_units=dense_units,
        dropout_rate=dropout_rate,
        learning_rate=learning_rate
    )

    early_stop = EarlyStopping(
        monitor='val_loss',
        patience=patience,
        restore_best_weights=True
    )

    history = model.fit(
        X_train,
        y_train,
        validation_split=0.2,
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[early_stop],
        verbose=1,
        shuffle=True
    )

    test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)

    print(f"\n=== {name} RESULTS ===")
    print("Test Accuracy:", test_acc)
    print("Best Validation Accuracy:", max(history.history['val_accuracy']))
    print("Epochs Trained:", len(history.history['loss']))

    return model, history, test_acc
```

```python
X_train, y_train = load_dataset(TRAIN_FILE)
X_test, y_test = load_dataset(TEST_FILE)

print("X_train:", X_train.shape, X_train.dtype)
print("y_train:", y_train.shape, y_train.dtype)
print("X_test :", X_test.shape, X_test.dtype)
print("y_test :", y_test.shape, y_test.dtype)
```

```python
# Baseline 3-layer CNN
baseline_model, baseline_history, baseline_acc = run_experiment(
    name="Baseline 3-Layer CNN",
    X_train=X_train,
    y_train=y_train,
    X_test=X_test,
    y_test=y_test,
    conv_filters=(32, 64, 64),
    dropout_rate=0.5,
    patience=5
)

# 1-layer CNN
one_layer_model, one_layer_history, one_layer_acc = run_experiment(
    name="1-Layer CNN",
    X_train=X_train,
    y_train=y_train,
    X_test=X_test,
    y_test=y_test,
    conv_filters=(32,),
    dropout_rate=0.5,
    patience=5
)

# Variant with last conv layer = 32 and dropout 0.55
variant_model, variant_history, variant_acc = run_experiment(
    name="3-Layer CNN Variant",
    X_train=X_train,
    y_train=y_train,
    X_test=X_test,
    y_test=y_test,
    conv_filters=(32, 64, 32),
    dropout_rate=0.55,
    patience=10
)

# Dropout 0.65
drop065_model, drop065_history, drop065_acc = run_experiment(
    name="3-Layer CNN Dropout 0.65",
    X_train=X_train,
    y_train=y_train,
    X_test=X_test,
    y_test=y_test,
    conv_filters=(32, 64, 64),
    dropout_rate=0.65,
    patience=5
)
# 3-layer CNN with dropout 0.60 and patience 12
drop060_pat12_model, drop060_pat12_history, drop060_pat12_acc = run_experiment(
    name="3-Layer CNN Dropout 0.60 Patience 12",
    X_train=X_train,
    y_train=y_train,
    X_test=X_test,
    y_test=y_test,
    conv_filters=(32, 64, 64),
    dropout_rate=0.60,
    patience=12
)
```

```python
results = {
    "Baseline 3-Layer CNN": baseline_acc,
    "1-Layer CNN": one_layer_acc,
    "3-Layer CNN Variant": variant_acc,
    "3-Layer CNN Dropout 0.65": drop065_acc,
    "3-Layer CNN Dropout 0.60 Patience 12": drop060_pat12_acc
}

print("\n=== ACCURACY SUMMARY ===")
for name, acc in results.items():
    print(f"{name}: {acc:.4f}")
```
